# Pie chart ( круговая диаграмма)

```python
stations = pd.read_csv(open('bus_stops.csv', 'r', encoding='cp1251'), sep=';')
```

```python
stations
```

```python
pavilion_count = stations.groupby(['Pavilion']).size()
```

```python
pavilion_count
```

```python
plt.pie(pavilion_count, labels=pavilion_count.index, autopct='%.2f')
plt.title('Bus stops pavilions')
plt.legend()
plt.show()
```

```python
admerea_count = stations.groupby(['AdmArea']).size()
```

```python
admerea_count
```

```python
labels=[]
for index in admerea_count.index:
    district = ''
    n=1
    index = index.split()
    for subname in index:
        district += subname[0].upper()
    while district in labels:
        district += index[0][n]
        n+=1
    labels.append(district)
labels
    
```

```python
plt.bar(labels, admerea_count)
plt.title('stations by districts')
plt.show()
```

## Подключение цветовых схем
(импорты должны быть в начале)

```python
from matplotlib.cm import get_cmap
```

```python
cmap = get_cmap('tab20')
cmap
```

```python
colors = [cmap(i) for i in range(len(labels))]
plt.bar(labels, admerea_count, color=colors)
plt.title('stations by districts')
plt.show()
```
