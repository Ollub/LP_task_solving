```python
import pandas as pd
```

```python
# changit output formating for floats
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2015 = pd.read_excel('russia_2015.xlsx', sheet_name='Partner')
export_2016 = pd.read_excel('russia_2016.xlsx', sheet_name='Partner')
```

```python
export_2015
```

```python
export_2016
```

```python

export = pd.concat([export_2015, export_2016])
# in this case dataframes have the same structure
# otherwise need to use additionals parametrs
```

```python
# random rows view
export.sample(10)
```

```python
export.pivot_table(index=['Year'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
export.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')
```

### Фильтрация по списку

```python
exclude = ['Europe & Central Asia', ' World', 'East Asia & Pacific', 'Middle East & North Africa',
          'United States Minor Outlying I', 'Unspecified']
# 1 item filter
# export_filtered = export[export['Partner Name'] != 'Europe & Central Asia']
# but for list
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], 
                   values='Export (US$ Thousand)', aggfunc='sum')\
                   .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

```python
export_pivot
```
